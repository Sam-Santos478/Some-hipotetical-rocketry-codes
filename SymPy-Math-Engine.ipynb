```python
import math as mth
import sympy as sp

while True:
    
    op = input("tell me the operation, if you don't want to calculate anything write 'N'")
    
    if  op.upper() == "N":
        break

    n1 = sp.sympify(input("Tell me the first number or your variable in case of you decided by diff or intt"))
    n2 = sp.sympify(input("Tell me the second number or your function related to your variable"))

    
    if n2 == 0 and op == "/" :
        display ("Serious? you realy want to make this heretic thing?")
        break
    elif op == "+":
        display(n1 + n2)
    elif op == "-":
        display(n1 - n2)
    elif op == "*":
        display(n1 * n2)
    elif op == "/":
        display (n1 / n2)
    elif op == "diff":
        deriv = sp.diff(n2, n1)
        display (deriv)
    elif op == "intt":
        integ = sp.integrate(n2, n1)
        display (integ)
```
