```python
# %%
import pandas as pd
import plotly.express as px

telemetriadados = pd.read_csv("rocket_telemetry_dataset.csv")
display (telemetriadados)
telemetriadados1 = telemetriadados[telemetriadados["Mission_ID"] == "M-001"]
telemetriadados2 = telemetriadados[telemetriadados ["Mission_ID"] == "M-002"]
telemetriadados3 = telemetriadados [telemetriadados ["Mission_ID"] == "M-003"]
display (telemetriadados1)
display (telemetriadados2)
display (telemetriadados3)
```

```python

# %%
display (telemetriadados.info())
telemetriadados = telemetriadados.dropna()
display(telemetriadados.info())
display(telemetriadados)

```

```python

# %%
display (telemetriadados.columns)
telemetriadados1 = telemetriadados1.sort_values (by= "Time", ascending=True)
telemetriadados2 = telemetriadados2.sort_values (by= "Time", ascending=True)
telemetriadados3 = telemetriadados3.sort_values (by= "Time", ascending=True)

telemetriadados1 = telemetriadados1.drop (columns = ["Flight_Type", "Mission_ID"])
telemetriadados2 = telemetriadados2.drop (columns = ["Flight_Type", "Mission_ID"])
telemetriadados3 = telemetriadados3.drop (columns = ["Flight_Type", "Mission_ID"])




for dado in telemetriadados1:
    if dado != "Time":
        gráficos1 = px.line(telemetriadados1, x= "Time", y= dado, title= "Telemetria de voo1")
        gráficos1.show()
display (telemetriadados1)


for dado in telemetriadados2:
    if dado != "Time":
        gráficos2 = px.line(telemetriadados2, x= "Time", y= dado, title= "Telemetria de voo2")
        gráficos2.show()
display (telemetriadados2)


for dado in telemetriadados3:
    if dado != "Time":
        gráficos3 = px.line(telemetriadados3, x= "Time", y= dado, title= "Telemetria de voo3")
        gráficos3.show()
display (telemetriadados3)

```

```python


# %%
# define a função com uma fonte de dados indefinida ainda e um texto compatível indefinido (df-ID, ID)
def telemetrycalc (df_ID, ID):
    print("Resultado:", ID)
    massainicial= df_ID["Mass"].max()
    massafinal = df_ID ["Mass"].min()

    gasto_propelente = massainicial - massafinal
    payload = massafinal
    print ("A carga payload adjunta do restante do hardware foi de", payload, "kg")
    print ("O propelente gasto em kg foi de", gasto_propelente, "kg")
    velocidade_max = df_ID ["Velocity"].max()
    print ("A máxima velocidade do foguete foi", velocidade_max, "m/s, ou", velocidade_max / 1000, "km/s")

    velocidade_min = df_ID ["Velocity"].min()

    deltav = velocidade_max - velocidade_min

    print ("O delta V foi de", deltav, "m/s")

    highest_temp = df_ID["Stag. Temp"].max()
    temp_celsius = (highest_temp - 32) * 0.555555556
    print (highest_temp, "Fahrenheit foi a máxima temperatura, em celsius equivale a:", temp_celsius)

    df_ID ["forcaG"] = df_ID["Accel"] / 9.81

    # display (df_ID) pra mostrar a tabela de cada base de dados separada anteriormente

    print("A máxima força G sofrida pelo hardware foi de:",df_ID["forcaG"].max())

    maxQ = df_ID["Dyn. Pressure"].max()
    print ("A máxima pressão aerodinâmica em pascais do voo do hardware foi", maxQ, "Pa")
    print("=====================================================================")
# todos os textos anteriores não definem a base de dados e o texto compativel delas, só definem a função a ser executada pelo python

telemetrycalc(telemetriadados1, "M-001")
telemetrycalc(telemetriadados2, "M-002")
telemetrycalc(telemetriadados3, "M-003")
# indica a base de dados e o texto compativel a ser executado na função pré-definida quantas vezes e por quantas bases de dados separadas você desejar
```
