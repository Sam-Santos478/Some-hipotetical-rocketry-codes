```python
import numpy as np

while True:

    def thrust_calc (df_ID, ID):
        print ("Analysing column:", ID)
        Max_Thrust = df_ID[ID].max()
        Min_Thrust = df_ID[ID].min()
        

        print("Your max Thrust was:", Max_Thrust, "Newtons")
        print("Your min Thrust was:", Min_Thrust, "Newtons")
    #redefine a sua base de dados relativo a variavel thrust_telemetry
    thrust_telemetry = pd.read_csv("thrust_curve.csv")
    #dá definição ao ID 
    thrust_calc(thrust_telemetry, "Empuxo")

    g = 9.81
    ISP = float(input("What is your engine ISP in seconds? Write right here:"))
    ve = g * ISP
    
    mass_flow = thrust_telemetry["Empuxo"] /ve
    grafic2 = px.line(x = thrust_telemetry["Tempo"], y = mass_flow, title = "Mass flow trough the flight in kg/s")
    grafic2.show()
    grafic1.show()
    #a função trapezoid integra o gráfico segundo os eixos que você seleciona pelo thrust_telemetry dividindo o gráfico em n trapezios e integrando
    It = np.trapezoid(y = thrust_telemetry["Empuxo"], x = thrust_telemetry["Tempo"])
    print("Your Total Impulse is:", It, "N*s")
    rerun =  input("If you want to rerun your calculator with another Thrust test data press 'S', if don't press 'N'.")
    if rerun.upper() == "N":
        break
     


```
