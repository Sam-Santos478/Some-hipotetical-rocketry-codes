```python
import math as mth
import time as tm

# manda rodar em loop
while True:
    
    print ("bem vindo a calculadora de Delta-V e altitude máxima de apogeu ideal!")
    m0 = float (input ("Diga-me a massa inicial de seu foguete em kg:"))
    tm.sleep (1)
    mf = float (input ("Ok, após a queima total de combustível deste estágio qual é a massa final do foguete em kg:"))
    tm.sleep (1)
    Isp = float(input ("Por fim, diga-me qual o impulso especifico do motor ou motores deste estágio em segundos:"))
    tm.sleep (1)

    deltaV = (Isp * 9.81) * mth.log(m0/mf)
    apogeu = (deltaV * deltaV) / (2 * 9.81)

    print("Seu Delta-V deste estágio especifico foi em metros por segundo igual a ", deltaV, " m/s.")
    print("Seu apogeu máximo ideal em metros é igual a ", apogeu, " m.")
    rerun = input("Deseja calcular outro ganho de Delta-V ou altitude? Digite 's', se não, digite 'n'")


    if rerun.lower() == "n":
        break 
```
